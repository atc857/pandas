# Importação de dados

Todos os formatos de importação (input) e exportação (output) do pandas

| Formato   | Input                 | Output            | Observação                     |
| --------- | --------------------- | ----------------- | ------------------------------ |
| CSV       | `pd.read_csv()`       | `.to_csv()`       | Arquivo Texto CSV, TSV, etc    |
| XLS/XLSX  | `pd.read_excel()`     | `.to_excel()`     | Planilha                       |
| HDF       | `pd.read_hdf()`       | `.to_hdf()`       | HDF5 database                  |
| SQL       | `pd.read_sql()`       | `.to_sql()`       | SQL table                      |
| JSON      | `pd.read_json()`      | `.to_json()`      | JavaScript Object Notation     |
| MSGPACK   | `pd.read_msgpack()`   | `.to_msgpack()`   | Portable binary format         |
| HTML      | `pd.read_html()`      | `.to_html()`      | código HTML                    |
| GBQ       | `pd.read_gbq()`       | `.to_gbq()`       | Google Big Query format        |
| DTA       | `pd.read_stata()`     | `.to_stata()`     | Stata                          |
| Parquet   | `pd.read_parquet()`   | `.to_parquet()`   | Apache Parquet                 |
| Feather   | `pd.read_feather()`   | `.to_feather()`   | Apache Arrow                   |
| Qualquer  | `pd.read_clipboard()` | `.to_clipboard()` | Ex., de pág HTML               |
| Qualquer  | `pd.read_pickle()`    | `.to_pickle()`    | (Structured) Python object     |

**Importando `CSV`**

Se atentar com os seguintes argumentos de [`pd.read_csv()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html):

* Caminho (`path`)
* `sep`: `','`, para europeu/brasileiro use `';'`
* `decimal`: `'.'`, para europeu/brasileiro use `','`
* `header`: `pandas` tenta adivinhar
* `index_col`: `None`, mas pode ser uma coluna do arquivo (ex: 2ª coluna use `index_col=2`)
* `names`: `None`, mas pode ser uma lista dos nomes das variáveis (colunas)
* `skip_rows`: `None` (pular linhas)
* `na_values`: `None`, mas pode ser qualquer string (ex: `'NA'`)
* `thousands`: `None` mas pode ser `','` ou `'.'`
* `encoding`
    - `'utf8'`: padrão
    - `'latin1'`: ç à é î ã

# Junção de *Dataframes*

Outra funcionalidade muito importante ao lidar com conjuntos de dados é juntar diferentes *dataframes*. Pandas oferece os métodos `join`, `merge` e `concat`. O `concat` é indicado para juntar as linhas de tabelas com colunas de mesmo nome. Por outro lado, `join` e `merge` servem para unir tabelas com base em  colunas-chaves. Essa é uma simplificação monstruosa. Para maiores detalhes, consulte a documentação.

| Opção    | Comportamento                                                                       |
| -------- | ----------------------------------------------------------------------------------- |
| `inner`  | Usa somente as combinações de chaves (*keys*) observadas em ambos *DataFrames*      |
| `left`   | Usa todas as combinações de chaves (*keys*) encontradas no *DataFrames* da esquerda |
| `right`  | Usa todas as combinações de chaves (*keys*) encontradas no *DataFrames* da direita  |
| `output` | Usa todas as combinações de chaves (*keys*) observadas em ambos *DataFrames* juntos |

![pandas-join](http://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png)

## Exemplo de concatenação de dois *dataframes* com as mesmas colunas

```python
# Para este exemplo, vamos criar um dataframe utilizando outra importante biblioteca do Python, a numpy.
import numpy as np
```

```python
# Gerando um conjunto de dados aleatórios
from numpy import random
questoes = {} # Cria um dicionário para receber o conjunto de dados
for i in range(10):
  questoes['q%s'%i] = random.randint(1,6,100)
```

```python
# Salvando o conjunto de respostas num dataframe e exibindo as primeiras linhas
respostas = pd.DataFrame.from_dict(questoes)
respostas.head()
```

```python
# Substituindo os valores de q0 para Masculino e Feminino
respostas['q0'].replace([1,2,3,4,5],['Masculino','Feminino','Masculino','Feminino','Masculino'], inplace=True)
# Removendo a coluna q5 e criando a coluna new_col
respostas.drop(['q5'], axis=1, inplace=True)
respostas['new_col'] = respostas['q2'] + respostas['q6']
respostas.head()
```

```python
respostas.info()
```

## Exemplo de junção de tabelas usando coluna-chave

**Merge**

![link text](https://pandas.pydata.org/pandas-docs/stable/_images/08_merge_left.svg)

Neste exemplo queremos utilizar os dados de população para comparar casos e mortes por covid entre diferentes países.

```python
# Os dados de covid estão no seguinte repositório
url_db = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/full_data.csv"
covid_db = pd.read_csv(url_db)
```

```python
covid_db.head()
```

```python
covid_db.tail()
```

Repare que neste *dataframe* da Our World in Data temos as quantidades totais de casos e mortes por covid para cada país (location). Queremos calcular casos e mortes pelo tamanho da população. Portanto, vamos baixar outro conjunto de dados com essa informação.

```python
url_pop = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/locations.csv'
pop = pd.read_csv(url_pop)
pop.head()
```

```python
# Desse dataframe interessa apenas a coluna-chave (location) e a population. Então, descarto as demais colunas
pop.drop(columns=['Country/Region', 'continent','population_year'], inplace=True)
pop.head()
```

```python
covid_db.shape
```

```python
pop.shape
```

```python
# Junção dos dois dataframes
df_merge = pd.merge(covid_db, pop)
```

```python
df_merge.head()
```

```python
df_merge.shape
```

## Lidando com dados faltantes (*missing values* `NaN`)

| Method         | Description                                                  |
| -------------- | ------------------------------------------------------------ |
| `df.dropna()`  | Filtra `NaN`, selecionar linhas (`axis = 0` é o padrão) ou selecionar colunas (`axis=1`) |
| `df.fillna()`  | Preencher os `NaN` com algum valor  |
| `df.isnull()`  | Retorna os valores booleanos indicando quais valores são `NaN` |

```python
# Eliminando as linhas com dados faltantes
df_merge.dropna(inplace=True)
df_merge.head()
```

```python
df_merge.shape
```

```python
# Criando as novas colunas calculadas
df_merge['total_cases_hab'] = df_merge['total_cases'] / df_merge['population'] * 1000000
df_merge['total_deaths_hab'] = df_merge['total_deaths'] / df_merge['population'] * 1000000
df_merge.head()
```

# Bônus

Seção com outras possibilidades de uso do Python para análise de dados.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# change date column to datetime format
df_merge['date'] = pd.to_datetime(df_merge['date'])
```

```python
#Total deaths per million inhabitants for countries with population above 10Mi of people
today = df_merge.date.iat[-1].strftime("%Y-%m-%d")
covid_deaths = df_merge.loc[(df_merge['population']>10000000)&(df_merge['date']==today)][['location','total_deaths_hab']].sort_values(by='total_deaths_hab', ascending=False).reset_index(level=0, drop=True)
covid_deaths.head(10)
```

```python
#Bar chart for total deaths per million inhabitants
deaths_10more = covid_deaths.iloc[:10,:].sort_values(by='total_deaths_hab')

plt.figure(dpi=150,figsize=(9,5))
plt.title('Total deaths per million inhabitants\n10 worst countries* in handling the covid-19 outbreak')
plt.text(6500,-0.7,today)
plt.text(1150,-0.7,'*Population over 10 million', fontsize=7)
plt.xlim([1000.0, 7500.0])
plt.xticks([])
clrs = ['darkgreen' if (x == 'Brazil') else 'darkred' for x in deaths_10more.location ]
plt.barh(deaths_10more.location , deaths_10more.total_deaths_hab, color=clrs)
for index, value in enumerate(deaths_10more.total_deaths_hab):
    plt.text(value, index, r'{0:.6}'.format(str(value)),va='center')
plt.subplots_adjust(bottom=0.015)
plt.figtext(0.125,-0.015,'Data: https://ourworldindata.org/coronavirus', fontsize=8)
plt.figtext(0.775,-0.015,'by André T. Campos', fontsize=8)
plt.show()
```
